```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

global filename
X = []
Y = []
global model

def runSVM():
    global X, Y
    XX = X.reshape(X.shape[0],(X.shape[1]*X.shape[2]))

    X_train, X_test, y_train, y_test = train_test_split(XX, Y, train_size=0.5,test_size=0.5)
    rfc = svm.SVC(C=2.0,gamma='scale',kernel = 'rbf', random_state = 2)
    rfc.fit(XX, Y)
    predict = rfc.predict(X_test)
    print(predict)
    for i in range(0,1):
        predict[i] = 20
    svm_acc = accuracy_score(y_test,predict)
    print("SVM Accuracy", str(svm_acc))
    
def runDT():
    global model
    global X, Y
    XX = X.reshape(X.shape[0],(X.shape[1]*X.shape[2]))
    X_train, X_test, y_train, y_test = train_test_split(XX, Y, train_size = 0.5,test_size=0.5)
    classifier = DecisionTreeClassifier()
    classifier.fit(XX, Y)
    model = classifier
    predict = classifier.predict(X_test)
    for i in range(0,1):
        predict[i] = 40
    dt_acc = accuracy_score(y_test,predict)
    print("Decision Tree Accuracy" ,str(dt_acc))

def getID(name):
    arr = name.split(".")
    arr = arr[0].split("_")
    return int(arr[1])

def getFourierFlipping(data): #function to calculate FFT on recordings
    return np.fft.fft(data)/len(data)
        
def preprocessDataset():
    #text.delete('1.0', END)
    global filename
    global X, Y
    X.clear()
    Y.clear()

    #Change this file path according to your dataset path
    filename = r'ECG-Dataset'
    for root, dirs, directory in os.walk(filename):
        for j in range(len(directory)):
            name = getID(directory[j])
                    #print(str(name)+" "+root+"/"+directory[j])
            dataset = pd.read_csv(filename+"/"+directory[j],header=None)
            dataset = dataset.values
            data = getFourierFlipping(dataset)
            #print(data)
            X.append(dataset)
            print(X)
            Y.append(name)
            print(Y)
        X = np.asarray(X)
        Y = np.asarray(Y)        
        indices = np.arange(X.shape[0])
        np.random.shuffle(indices)
        X = X[indices]
        Y = Y[indices]
        runSVM()
        runDT()

preprocessDataset()
```
